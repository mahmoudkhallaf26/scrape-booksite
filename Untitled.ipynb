```python
import csv
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
new_link = "http://books.toscrape.com/catalogue/category/books/mystery_3/index.html"
```

```python
new_response = requests.get(new_link)
```

```python
soup = BeautifulSoup(new_response.content,'html.parser')
soup
```

```python
h3 = soup.find_all("h3")
h3
```

```python
name_of_book =[]
for name in range(len(h3)):
    name_of_book.append(h3[name].find_all("a")[0].get("title"))
```

```python
print(name_of_book)
```

```python
prices = soup.find_all("p",attrs={"class":"price_color"})
```

```python
list_of_prices=[]
for x in range(len(prices)):
    price=prices[x].text
    list_of_prices.append(price)
```

```python
list_of_prices
```

```python
rate = soup.find_all("p",attrs={"class":"star-rating"})
```

```python
star_list = []
for i in range(len(rate)):
    star=rate[i].get_attribute_list("class")
    star_list.append(star[1])
```

```python
star_list
```

```python
with open("scrape_book","w",newline="") as f:
    writer = csv.DictWriter(f,fieldnames=["Book_name","price","rate"])
    writer.writeheader()
    for i in range(len(name_of_book)):
        book_name = name_of_book[i]
        price = list_of_prices[i]
        rate = star_list[i]
        writer.writerow({"Book_name":book_name,"price":price,"rate":rate})
```
